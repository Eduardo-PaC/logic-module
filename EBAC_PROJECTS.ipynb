#CALCULADORA AUTOMÁTICA - 01 PASSO PARA O FUTURO

```python
# Solicita o nome do usuário do programa.
nome_u = input("Olá, Usuário. Por favor, digite o seu nome: ")
print(f"Olá, {nome_u}. Seja bem-vindo à sua calculadora inteligente!")

operacoes = {
    "+": lambda x, y: x + y,
    "-": lambda x, y: x - y,
    "*": lambda x, y: x * y,
    "/": lambda x, y: x / y,
}

while True: #Cria um loop infinito

  # Pergunta se o usuário deseja utilizar o programa ou não
  escolha = input("Você deseja Continuar? Digite Sim para prosseguir e Não para sair. Resposta do Usuário: ")
  if escolha == "Não":
    print("Finalizando aplicação do sistema. Obrigado por usar nosso programa!")
    break #Finaliza o loop

  # Define o primeiro número
  n1 = float(input("Agora vamos começar, digite o primeiro número: "))

  # Define a operação matemática
  op = input("Agora digite a operação desejada (+, -, *, /): ")

  if op not in operacoes:
    print("Operação inválida")
    continue

  # Define o segundo número
  n2 = float(input("Agora digite o segundo número: "))

  # Traz o resultado da operação

  resultado = operacoes[op](n1, n2)

  print(f"Perfeito, o resultado da sua operação é {resultado: .0f}")
```
