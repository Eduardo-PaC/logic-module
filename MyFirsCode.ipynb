##Tabela Verdade

And

```python
print(25 > 2 and 2 < 5)   # V and V = V
print(25 > 2 and 2 > 5)   # V and F = F
print(25 < 2 and 2 < 5)   # F and V = F
print(25 < 2 and 2 > 5)   # F and F = F
```

Or

```python
print(5 > 3 or 3 < 4)   # V or V = V
print(5 > 3 or 3 > 4)   # V or F = F
print(5 < 3 or 3 < 4)   # F or V = F
print(5 < 3 or 3 > 4) # F or F = F
```

Not

```python
not 10 > 2   # V = F
```

```python
not 10 < 2   # F = V
```

```python
# If / Elif / Else
num1 = input("Número do seu campeão p1: ")
num2 = input("Número do seu campeão p2: ")

if num1 > num2:
  print(num1, "Você é demaaais")

elif num1 < num2:
  print(num2, "Você é imbatível")

else:
  print("Ambos são uns franguinhos")
```

Condicionais com Operadores Lógicos

```python
# If / Elif / Else / And

n1 = int(input("Digite seu campeão p1: "))
n2 = int(input("Digite seu campeão p2: "))
n3 = int(input("Digite seu campeão p3: "))

if (n1 > n2) and (n1 > n3):
  print(n1, "Você é uma MÁQUINAAAA")

elif (n2 > 1) and (n2 > n3):
  print(n2, "Você é o Soberano!!!")

else:
  print("Parabéns", n3, "Você conquistou o campo")
```

```python
nome = input("Digite o nome do aluno sendo avaliado: ")
nota = float(input("Digite a Nota do aluno: "))
média = float(8)
greencard = input("Digite se o aluno possuí Green Card ") # Passe livre por boa conduta e méritos acadêmicos admiráveis.

if nota >= média and greencard == "Sim":
  print("Aluno", nome, "Aprovado com Honras")

elif nota >= média and greencard == "Não":
  print("Aluno", nome, "Aprovado")

elif nota >= média or greencard == "Sim":
  print("Aluno", nome, "Aprovado")

else:
  print("Aluno", nome, "Reprovado")
```
